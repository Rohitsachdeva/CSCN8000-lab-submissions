### **Predict Canada’s per capita income in the year 2020. There is a Canada_per_capita_income.csv file. Using this, build a regression model and predict the per capita income for canadian citizens in the year 2020**

## **Rohit Kumar 8895806**

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
```

```python
data = pd.read_csv("C:/Users/Rohit/Downloads/canada_per_capita_income.csv")
```

```python
# Create the linear regression model
model = LinearRegression()

# Prepare the data
X = data[['year']]
X = X.values
y = data['per capita income (US$)']

# Fit the model to the data
model.fit(X, y)
```

```python
# Predict the per capita income for 2020

predicted_income_2020 = model.predict([ [2020] ])

# Print the predicted income
print("Predicted per capita income in 2020:", predicted_income_2020[0])
```

```python
import matplotlib.pyplot as plt


# Plot the data points and the regression line
plt.scatter(X, y, color='blue', label='Actual Data')
plt.plot(X, model.predict(X), color='red', label='Regression Line')

# Plot the predicted income for 2020
plt.scatter(2020, predicted_income_2020, color='green', label='Predicted Income (2020)')

# Set the plot labels and title
plt.xlabel('Year')
plt.ylabel('Per Capita Income (US$)')
plt.title('Canada Per Capita Income Over Time')

# Add a legend
plt.legend()

# Display the plot
plt.show()
```
